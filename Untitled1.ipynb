```python
import tornado.web
import tornado.httpserver
import tornado.ioloop
import torndsession
from tornado import gen
import tornado.httpclient
from torndsession.sessionhandler import SessionBaseHandler
import random
from random import randrange
from pyroaring import BitMap
from datetime import datetime
from datetime import timedelta
import time


def expiration_date(start = datetime.strptime("7/1/2020 1:30 PM", "%m/%d/%Y %I:%M %p"), end= datetime.now()):                   
    difference = end -start
    seconds = (difference.days * 24 * 60 * 60) + difference.seconds
    random_second = randrange(seconds)
    random_day = randrange(30)
    rand_date = start + timedelta(days=random_day,seconds=random_second)
    if (rand_date - datetime.now()).total_seconds() > 0:
        rand_date = start + timedelta(seconds=random_second)
        return rand_date
    else:
        return rand_date

def last_week_visit(past_date):    
    now = datetime.now()
    past = past_date 
    difference = (past - now).total_seconds()
    
    return abs(difference /(60 * 60 * 24)) <= 7


class Application(tornado.web.Application):
    def __init__(self):
        handlers = [
            (r"/", IntroHandler),
            (r"/([0-9]+)", MainHandler),
            (r"/ConTest1000", ConHandler)
        ]
        settings = dict(debug=True),        
        tornado.web.Application.__init__(self, handlers)

        
class MainHandler(tornado.web.RequestHandler):

    def get(self, id):
        if int(id) > 100:
            id = 100
            self.write("Maximum number of sessions was set to {}\n\n".format(id))
            
        for y in range(1, 11):
            globals()["tag%s" % y] = BitMap()        
        weekly_visit = BitMap()
        
        self.write("-----------------------------------------------------------\n")
        self.write("Show the first 10 Users\n")
        self.write("-----------------------------------------------------------\n\n")
        for i in range(1, int(id)+1):
            setter = randrange(10)
            if setter == 0:
                tag1.add(i)
            elif setter == 1:
                tag2.add(i)
            else:
                pass

            session_time = expiration_date()            
            visited = last_week_visit(session_time)
            if visited: weekly_visit.add(i)                
            session_time = session_time.isoformat()
            
            setter = "tag_" + str(setter+1) + "_" + session_time
            session = self.set_cookie(name = str(i), value = setter)  #,expires = expiration_date())
            if i <= 10: self.write("The {} cookie was set with :  {}\n".format(i, setter))
        self.write("\n-----------------------------------------------------------")
        self.write("\n\n>>  Total Number of Users are: {}\n".format(id))
        self.write("\n>>  Number of Users have NOT logged in for a week: {}\n".format(len(list(weekly_visit))))
        self.write("\n>>  Number of Users with tag1 and tag2 are: {}\n".format(len(list(tag1))+ len(list(tag2))))
        self.write("\n>>  The User ID's with tag1 are below:\n\n")
        self.finish({'tag1': list(tag1)})
                    

# Concurrent user testing measures how long it takes the server
# to respond to a specified number of simultaneous requests
class ConHandler(tornado.web.RequestHandler):    
    
    @gen.coroutine
    def get(self):
        start = time.time()
        for i in range(1000):            
            http_client = tornado.httpclient.AsyncHTTPClient()
            response = yield http_client.fetch("http://localhost:8002")
        total_time = time.time()-start
        self.write("Done!!\nFetched URL 1000 times\n\n")
        self.write("-----------------------------------------------------------\n")
        self.write("Total time elapsed: {:.6f} seconds\n".format(total_time))
        self.write("Time per request: {:.8f} seconds\n".format(total_time/1000))

class IntroHandler(tornado.web.RequestHandler):
    def get(self):
        self.render("index.html")

        
def main():
    http_server = tornado.httpserver.HTTPServer(Application())
    http_server.listen(8002)
    tornado.ioloop.IOLoop.instance().start()

if __name__ == "__main__":
    main()
```
